1.Breadth-First-Search(BFS)

```python
from collections import deque
graph = {
    "MVLU College": ["Marol Naka"],
    "Marol Naka": ["Aarey Check Naka"],
    "Aarey Check Naka": ["Aarey Unit 22"],
    "Aarey Unit 22": []
}
cost = {
    ("MVLU College", "Marol Naka"): 2,
    ("Marol Naka", "Aarey Check Naka"): 3,
    ("Aarey Check Naka", "Aarey Unit 22"): 2
}
start = "MVLU College"
goal = "Aarey Unit 22"
queue = deque([(start, [start], 0)])
visited = set()
nodes = 0
while queue:
    node, path, total_cost = queue.popleft()
    nodes += 1

    if node == goal:
        print("Path:", " -> ".join(path))
        print("Total Cost:", total_cost, "km")
        print("Nodes Explored:", nodes)
        break
    visited.add(node)
    for neighbor in graph[node]:
        if neighbor not in visited:
            queue.append((neighbor, path + [neighbor],
                          total_cost + cost[(node, neighbor)]))
```

2. Depth First Search (DFS)

```python
graph = {
    "MVLU College": ["Marol Naka"],
    "Marol Naka": ["Aarey Check Naka"],
    "Aarey Check Naka": ["Aarey Unit 22"],
    "Aarey Unit 22": []
}
cost = {
    ("MVLU College", "Marol Naka"): 2,
    ("Marol Naka", "Aarey Check Naka"): 3,
    ("Aarey Check Naka", "Aarey Unit 22"): 2
}
stack = [("MVLU College", ["MVLU College"], 0)]
visited = set()
nodes = 0
while stack:
    node, path, total_cost = stack.pop()
    nodes += 1
    if node == "Aarey Unit 22":
        print("Path:", " -> ".join(path))
        print("Total Cost:", total_cost, "km")
        print("Nodes Explored:", nodes)
        break
    if node not in visited:
        visited.add(node)
        for neighbor in reversed(graph[node]):
            stack.append((neighbor, path + [neighbor],
                          total_cost + cost[(node, neighbor)]))
```

3. Uniform Cost Search (UCS)

```python
import heapq
graph = {
    "MVLU College": [("Marol Naka", 2)],
    "Marol Naka": [("Aarey Check Naka", 3)],
    "Aarey Check Naka": [("Aarey Unit 22", 2)],
    "Aarey Unit 22": []
}
pq = [(0, "MVLU College", ["MVLU College"])]
visited = {}
nodes = 0
while pq:
    total_cost, node, path = heapq.heappop(pq)
    nodes += 1
    if node == "Aarey Unit 22":
        print("Path:", " -> ".join(path))
        print("Total Cost:", total_cost, "km")
        print("Nodes Explored:", nodes)
        break
    if node not in visited or total_cost < visited[node]:
        visited[node] = total_cost
        for neighbor, cost in graph[node]:
            heapq.heappush(pq, (total_cost + cost,
                                neighbor,
                                path + [neighbor]))
```

5. A* Search

```python
import heapq
heuristic = {
    "MVLU College": 7,
    "Marol Naka": 5,
    "Aarey Check Naka": 2,
    "Aarey Unit 22": 0
}
graph = {
    "MVLU College": [("Marol Naka", 2)],
    "Marol Naka": [("Aarey Check Naka", 3)],
    "Aarey Check Naka": [("Aarey Unit 22", 2)],
    "Aarey Unit 22": []
}
pq = [(7, 0, "MVLU College", ["MVLU College"])]
visited = {}
nodes = 0
while pq:
    f, g, node, path = heapq.heappop(pq)
    nodes += 1
    if node == "Aarey Unit 22":
        print("Path:", " -> ".join(path))
        print("Total Cost:", g, "km")
        print("Nodes Explored:", nodes)
        break
    if node not in visited or g < visited[node]:
        visited[node] = g
        for neighbor, cost in graph[node]:
            new_g = g + cost
            new_f = new_g + heuristic[neighbor]
            heapq.heappush(pq, (new_f,
                                new_g,
                                neighbor,
                                path + [neighbor]))
```
